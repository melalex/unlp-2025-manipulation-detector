```python
import pandas as pd


data = {
    'Model': [
        'bert-base-multilingual-cased',
        'all-MiniLM-L6-v2',
        '(RU)rubert-base-cased-conversational',
        '(UA)ukr-roberta-base',
        'xlm-roberta-base',
        'rubert-base-cased-conversational'
    ],
    'Score': [
        0.495283,
        0.425846,
        0.549884,
        0.544135,
        0.49,
        0.546779
    ]
}

# Створення DataFrame
df = pd.DataFrame(data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Налаштування стилю seaborn
sns.set_theme(style="darkgrid")

# Створення графіка
ax = sns.barplot(x='Model', y='Score', hue='Model', data=df, palette='viridis', legend=False)

# Додавання підписів значень
for p in ax.patches:
    ax.annotate(f'{p.get_height():.4f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()/2), 
                ha='center', va='bottom', 
                color='black', 
                fontsize=12)

# Налаштування заголовка та підписів осей
plt.title('F1 Scores', fontsize=16)
plt.xticks(rotation=45, ha='right')

# Видалення підписів осей
ax.set_xlabel('')
ax.set_ylabel('')

# Показати графік
plt.tight_layout()
plt.show()
```
